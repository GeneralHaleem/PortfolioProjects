# Stack Overflow Developer Survey

 Each year, more than 70,000 developers share with stack overflow their learning and advancement strategies, tool preferences, and goals. Over 180 nations responded to the yearly developers survey, which examines all facets of the developer experience from learning to code to their preferred technology. 
 
 The purpose of this project is to analyze the data to find great insights about the attitudes, tools, and environments that are shaping the art and practice of software today.
