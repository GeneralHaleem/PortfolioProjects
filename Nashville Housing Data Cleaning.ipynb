# Nashville Housing

Importing Necessary Libraries

```python
import pandas as pd
```

Importing Dataset

```python
pd.set_option("display.max_rows", None)
```

```python
Nashville_Housing = pd.read_csv("Nashville Housing.csv")
Nashville_Housing.head()
```

Optimization

```python
Nashville_Housing.info()
```

```python
Nashville_Housing.nunique()
```

```python
Nashville_Housing["LandUse"] = Nashville_Housing["LandUse"].astype("category")
Nashville_Housing["SoldAsVacant"] = Nashville_Housing["SoldAsVacant"].astype("category")
Nashville_Housing["TaxDistrict"] = Nashville_Housing["TaxDistrict"].astype("category")
Nashville_Housing["Bedrooms"] = Nashville_Housing["Bedrooms"].astype("category")
Nashville_Housing["FullBath"] = Nashville_Housing["FullBath"].astype("category")
Nashville_Housing["HalfBath"] = Nashville_Housing["HalfBath"].astype("category")
```

```python
Nashville_Housing.info()
```

Standardizing Date Formart

```python
Nashville_Housing["SaleDate"] = pd.to_datetime(Nashville_Housing["SaleDate"])
Nashville_Housing.head()
```

Working on duplicated values

```python
Nashville_Housing.duplicated().sum()
```

Working on Null Values

```python
Nashville_Housing.isnull().sum()
```

Splitting PropertyAddress Column

```python
Property_Address = Nashville_Housing["PropertyAddress"].str.split(",", expand= True, n = 2)
Property_Address = Property_Address[0]
Property_Address.head()
```

```python
Nashville_Housing.insert(loc = 4, column = "Property_Address", value = Property_Address)
Nashville_Housing.head()
```

```python
Property_Address = Nashville_Housing["PropertyAddress"].str.split(",", expand= True, n = 2)
Property_City = Property_Address[1]
Property_City.head()
```

```python
Nashville_Housing.insert(loc = 5, column = "Property_City", value = Property_City)
Nashville_Housing.head()
```

Splitting Owners Address

```python
Owner_Address = Nashville_Housing["OwnerAddress"].str.split(',', expand = True, n= 3)
Owner__Address = Owner_Address[0]
Owner__Address.head()
```

```python
Nashville_Housing.insert(loc = 12, column = "Owner_Address", value = Owner__Address)
Nashville_Housing.head()
```

```python
Owner_Address = Nashville_Housing["OwnerAddress"].str.split(',', expand = True, n= 3)
Owner_City = Owner_Address[1]
Owner_City.head()
```

```python
Nashville_Housing.insert(loc = 13, column = "Owner_City", value = Owner_City)
Nashville_Housing.head()
```

```python
Owner_Address = Nashville_Housing["OwnerAddress"].str.split(',', expand = True, n= 3)
Owner_Country = Owner_Address[2]
Owner_Country.head()
```

```python
Nashville_Housing.insert(loc = 14, column = 'Owner_Country', value = Owner_Country)
Nashville_Housing.head()
```

Replacing Yes with Y and No with N in the SoldAsVacant column

```python
Nashville_Housing["SoldAsVacant"] = Nashville_Housing["SoldAsVacant"].str.replace("No", "N")
Nashville_Housing["SoldAsVacant"] = Nashville_Housing["SoldAsVacant"].str.replace("Yes", "Y")
Nashville_Housing["SoldAsVacant"].head()
```

Deleting Unused Columns

```python
Nashville_Housing = Nashville_Housing.drop(columns = ["PropertyAddress", "OwnerAddress"])
Nashville_Housing.head()
```

Re-Spelling wrongly spelt columns

```python
Nashville_Housing.rename(columns = {"Acreage": "Average"})
```
